```python
import tensorflow as tf
tf.__version__
```

```python
tf.random.set_seed(42)
tensor = tf.constant(tf.random.uniform(shape=[50]), shape = [1, 1, 1, 1, 50])
tensor
```

```python
" docs: Removes dimensions of size 1 from the shape of a tensor. "
tf.squeeze(tensor)
```

```python
tensor = tf.constant(tf.random.uniform(shape=[50]), shape = [1, 1, 25, 2, 1])
tensor
```

```python
tf.squeeze(tensor) # removes left 1 and right 1 from shape
```
